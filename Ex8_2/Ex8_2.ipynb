```python
import math
import numpy as np
import matplotlib.pyplot as plt

start = -2 * math.pi                             
stop = 2 * math.pi                               
step = (2 * math.pi - -2 * math.pi) / (500 - 1)  
x = [loop1 for loop1 in np.arange(start, stop+step, step)]    
y = [math.cos(x)**2 for x in x]                               
plt.plot(x,y)                                    
plt.show()


x = np.linspace(start,stop,500)    
y = np.cos(x)**2                  
plt.plot(x,y)                      
plt.show() 
```
